```python
with open("input.txt","r") as f:
    l=[]
    l2=[]
    keywords=['if', 'else', 'while', 'for', 'int', 'float', 'bool', 'true', 'false']
    k=[]
    operator=['-','+','=','*','/','%','**','//']
    o=[]
    logicalop=['>','<','&&','||','!']
    lo=[]
    numerics = []
    identifiers=[]
    literal=[]
    others=[]
    for line in f:
        l.append(line.split())
    for i in l:
        i2=[]
        for j in i:
            a=j
            if a[-1]==',' or a[-1]==';' or a[-1]==')':
                j=a[0:-1]
                i2.append(j)
                if a[-1]==',':
                    i2.append(',')
                if a[-1]==')':
                    i2.append(')')
                else:
                    i2.append(';') 
            else:
                i2.append(j)
        l2.append(i2)        
        
    for i in l2:
        for j in i:
            if j in keywords:
                if j not in k: 
                    k.append(j)
            elif j in operator:
                if j not in o:
                    o.append(j)
            elif j in logicalop:
                if j not in lo:
                    lo.append(j)
            elif j[0].isdigit():
                numerics.append(j)
            elif (ord(j[0])>= 97 and ord(j[0])<=122) or (ord(j[0])>= 65 and ord(j[0])<=90):
                if j not in keywords:
                    if j not in identifiers:
                        identifiers.append(j)
            elif j[0]=='"' and j[-1]=='"':
                if j not in literal:
                    literal.append(j)
            else:
                if j not in others:
                    others.append(j)
print("List of all the tokens")
print("Keywords: ")
for i in k:
    print(i)  
print("Identifiers: ")
for i in identifiers:
    print(i) 
print("Math Operators: ")
for i in o:
    print(i)   
print("Logical Operators: ")
for i in lo:
    print(i)
print("Numerical Values: ")
for i in numerics:
    print(i)
print("Literals: ")
for i in literal:
    print(i)
print("Others: ")
for i in others:
    print(i)   
```
